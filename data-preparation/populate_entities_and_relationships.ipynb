```python
from database.database import Database

db = Database()
```

```python
from llm_service.gemini_service import generate_subgraph
import time

db_response = db.get_all_pdf_documents()
pdfs = db_response.data

# For each entry pdf, create entities and relationships
for pdf in pdfs:
    try:
        time.sleep(6)
        
        subgraph = generate_subgraph(pdf['raw_content'])
        entities = subgraph['entities']
        relationships = subgraph['relationships']

        print("="*50)
        print("length of entities", len(entities))
        print("length of relationships", len(relationships))

        entity_ids = {}

        for entity in entities:
            entity_entry = db.create_entity(entity['natural_id'], entity['title'], entity['type'], entity['attributes'], [pdf['doc_id']])
            entity_ids[entity['natural_id']] = entity_entry.data[0]['entity_id']

        for relationship in relationships:
            source_id = entity_ids[relationship['source']]
            target_id = entity_ids[relationship['target']]
            db.create_relationship(source_id, target_id, relationship['label'], relationship['attributes'], [pdf['doc_id']])

        print(f"pdf with id {pdf['doc_id']} processed successfully")
            
    except Exception as e:
        print(f"pdf with id {pdf['doc_id']} failed to be processed: {e}")
        continue
```
